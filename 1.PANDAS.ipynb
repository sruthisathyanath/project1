# Cleaning Data

```python
df3=pd.read_csv('/home/admin1/LUMINAR/datasource/dirtydata.csv') #to load a csv file into df3
```

```python
print(df3)
```

```python
newdf3=df3 #copying into a new dataframe 
```

```python
print(newdf3)
```

```python
#METHOD 1 - EMPTY CELLS

new_df=newdf3.dropna()  # to treat empty values  
```

```python
print(new_df)  # row 18,22,28 is removed since it contain nan values,  print(new_df.to_string())
```

```python
newdf3 # no changes in newdf3, ie row 18,22,28 is not removed 
       # if we do newdf3.dropna(inplace=True)  this will remove 18,22,28 rows once and for all
       
# if we do newdf=newdf3.dropna(inplace=True) and print newdf the output willbe none
       
# Reason:This is because function with inplace=True does not return anything. 
        
#It modifies an existing data frame with the desired operation and does it ‘in place’, that is on the original data frame.
```

```python
#METHOD 2
newdf3.fillna(130)                  #replaces all empty cells with value 130 in the whole Data Frame.


#newdf3.fillna(130, inplace = True) ==> to change NaN values to 130 for ones and for all
```

```python
#replace empty values for one column, specify the column name

d1=newdf3
d1["Calories"].fillna(130)  #replacing empty values in calories column with 130


#d1["Calories"].fillna(130, inplace = True) ==> changing NaN values to 130 for ones and all
```

```python
dfcor= pd.read_csv('/home/admin1/LUMINAR/datasource/datacor.csv')
```
